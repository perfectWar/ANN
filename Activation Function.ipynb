# Activation function

```python
import numpy as np
```

```python
x = np.linspace(-10,10)
print(x)
```

```python
y = np.where(x>1,1,0)
```

```python
import matplotlib.pyplot as plt
```

```python
y
```

```python
x
```

```python
plt.plot(x,y)
plt.title("Binary Step activation function")
plt.xlabel("Net input")
plt.ylabel("Output")
```

```python
def linearActivation(x):
    y = x
    return y
x = np.linspace(-10,10)
plt.plot(x,linearActivation(x))
plt.title("Linear Activation Function")
plt.xlabel("Net input")
plt.ylabel("Output")
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
x = np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.title("Sigmoid Function")
plt.xlabel("Net input")
plt.ylabel("Output")
```

```python
def tanhyperbolic(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
x = np.linspace(-10, 10) 
plt.plot(x, tanhyperbolic(x))
plt.title("Tanhyperbolic Function")
plt.xlabel("Net input")
plt.ylabel("Output")
```

```python
def Rel(x):
    return np.where(x>=0,x,0)
x = np.linspace(-10, 10) 
plt.plot(x, Rel(x))
plt.title("Rel Function")
plt.xlabel("Net input")
plt.ylabel("Output")
```

```python
def softmax(x):
    return (np.exp(x))/(np.sum(np.exp(x),axis=0))
x = np.linspace(-10, 10) 
plt.plot(x, softmax(x))
plt.title("Softmax Function")
plt.xlabel("Net input")
plt.ylabel("Output")
```
